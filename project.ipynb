```python
import re
import string
from collections import Counter
import numpy as np
```

```python
def read_file(filename):
    with open(filename, encoding='utf8') as file:
        lines = file.read()
        words = []
        for line in lines.split():
            print(line)
            words.append(line)
    return words
```

```python
words = read_file('./words.txt')
print(words)
print(f"There are {len(words)} total words in the file")
```

```python
vocabs = set(words)
print(f"There are {len(vocabs)} unique words in the vocabulary")
```

```python
words_count = Counter(words)
print(words_count['Негізгі'])

print(words_count)
```

```python
total_word_count = float(sum(words_count.values()))
print(total_word_count)
print(words_count.values())
print(words_count['Негізгі'])
word_probas = { word: words_count[word] / total_word_count for word in words_count.keys() }
```

```python
print(word_probas['негізгі'])
```

```python
def split(word):
    return [(word[:i], word[i:]) for i in range(len(word) + 1)]

def delete(word):
  return [l + r[1:] for l,r in split(word) if r]

def swap(word):
  return [l + r[1] + r[0] + r[2:] for l, r in split(word) if len(r)>1]

def replace(word):
  letters = 'аәбвгғджзикқлмнңоөпрстұүфхцчщъьыіэюя'
  return [l + c + r[1:] for l, r in split(word) if r for c in letters]

def insert(word):
  letters = 'аәбвгғджзикқлмнңоөпрстұүфхцчщъьыіэюя'
  return [l + c + r for l, r in split(word) for c in letters]
```

```python
print(replace("Түлкі"))
```

```python
print(insert("Түлкі"))
```

```python
def edit1(word):
  return set(delete(word) + swap(word) + replace(word) + insert(word))
```

```python
print(edit1("Түлкі"))
```

```python
def edit2(word):
  return set(e2 for e1 in edit1(word) for e2 in edit1(e1))
```

```python
print(edit2("Түлкі"))
```

```python
def check_word(word, vocabulary, word_probab):
    if word in vocabulary:
        print(f"{word} is already correct")
        return
    suggestions = edit1(word) or edit2(word) or [word]
    best_guesses = [w for w in suggestions if w in vocabulary]
    return [(w, word_probab[w]) for w in best_guesses]
```

```python
word1 = 'Ғегізгі'
corrections = check_word(word1, vocabs, word_probas)

if corrections:
    probs = np.array([c[1] for c in corrections])
    index = np.argmax(probs)
    correct = corrections[index][0]
    print(f"{correct} ====> {word1}")
```

```python
word = 'тулки'

outputs = spell.lookup(word, Verbosity.CLOSEST, max_edit_distance = 2)

for output in outputs:
    print(output)
```
